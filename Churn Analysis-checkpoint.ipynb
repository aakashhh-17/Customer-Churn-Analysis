# Churn Analysis of customers in telecom industry

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Data analyst projects\churn analysis\Customer Churn.csv")
df.head()
```

```python
df.info()
```

### Replacing blanks with 0 as tenure is 0 and no charges are recorded

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", 0)
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

### Converted value of 0 and 1 of senior citizen to No and Yes to make it easier to understand. 

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype('string')
df['SeniorCitizen'] = df['SeniorCitizen'].str.replace('0', 'No')
df['SeniorCitizen'] = df['SeniorCitizen'].str.replace('1', 'Yes')
```

```python
df
```

```python
ax = sns.countplot(x = 'Churn', data =  df) 
ax.bar_label(ax.containers[0])
plt.title('Count of customers by churn')
plt.figure(figsize = (4,4))
plt.show()
```

```python
plt.figure(figsize = (5,5))

gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct = '%1.2f%%')
plt.title('Percentage of churned customer')
plt.show()
```

### From the given pie chart we can conclude that 26.54% customers are churned out.
### Now lets explore the reasons behind it.

```python
plt.figure(figsize = (4,4))
gn = sns.countplot(x = df['gender'], data = df, hue = 'Churn')
gn.bar_label(gn.containers[0])
plt.title('Churn by gender')
plt.show()
```

```python
plt.figure(figsize = (4,4))
gn = sns.countplot(x = df['SeniorCitizen'], data = df, hue = 'Churn')
gn.bar_label(gn.containers[0])
plt.title('Churn by SeniorCitizen')
plt.show()
```

```python
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Calculate percentages of total
grouped_perc = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = grouped_perc.plot(kind='bar', stacked=True, figsize=(6,6), colormap='viridis')

# Add percentages on the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%', label_type='center')

plt.title('Churn by SeniorCitizen (as % of total)')
plt.ylabel('Percentage')
plt.xlabel('SeniorCitizen')
plt.show()
```

### Comparatively a greater percentage of senior citizens have churned out.

```python
plt.figure(figsize = (9,5))
sns.histplot(x = 'tenure', data = df, bins = 72, hue = 'Churn')
plt.show()
```

### People who have used our services for long time have stayed and those who churned have only used for initial months.

```python
plt.figure(figsize = (4,4))
gn = sns.countplot(x = df['Contract'], data = df , hue = 'Churn')
gn.bar_label(gn.containers[0])
plt.title('Count of customers by contract')
plt.show()
```

### People who have month-to-month contract are likely to churn comparatively by those who have one year or two year contracts. 

```python
df.columns
```

```python
# Define the columns
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set the number of subplots (3 rows, 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through the columns and create a countplot for each
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df['Churn'])
    axes[i].set_title(f'Count of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

### Across most service categories, a higher proportion of customers who have not churned is observed (blue bars). For services like OnlineSecurity, TechSupport, and DeviceProtection, customers who opted out of these services (No) exhibit a higher churn rate (orange bars), indicating these services may be linked to customer retention.

```python
plt.figure(figsize = (7,5))
gn = sns.countplot(x = df['PaymentMethod'], data = df , hue = 'Churn')
gn.bar_label(gn.containers[0])
gn.bar_label(gn.containers[1])
plt.title('Churn customers by PaymentMethod')
plt.xticks(rotation = 45)
plt.show()
```
